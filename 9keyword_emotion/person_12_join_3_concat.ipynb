##### 사람문장1, 사람문장2를 새로운 컬럼(사람문장)에 넣기

```python
new_df_merge = df_drop_panic
new_df_merge['사람문장'] = data_df[['사람문장1','사람문장2']].apply(' '.join, axis = 1)
```

```python
data1 = new_df_merge.drop(['사람문장1','사람문장2','사람문장3'],axis = 1)
data1
```

```python
data2 = new_df_merge.drop(['사람문장1','사람문장2'],axis = 1)
data2
```

```python
data_2 = data2.dropna()
data_2
```

```python
data_3 = data_2.rename(columns = {'사람문장3':'사람문장','사람문장':'삭제예정'})
data_3
```

```python
df_data=data_3.drop(['삭제예정'],axis = 1)
df_data
```

```python
# 열이름이 같으면 지정안해도 알아서 붙음
dt = pd.concat([data1,df_data])
dt
```

```python
f_data = dt[['상황키워드','감정_대분류', '사람문장']]
f_data
```
